```python
data.drop(columns=['eskomExports','eskomImports', 'total_Exports', 'total_Imports'], inplace=True)
```

```python
sns.boxplot(data.drop(columns=['year', 'quarter']),
            orient='h')
plt.xticks(rotation=90);
```

```python
q1_unadj_produced = data["unadjustedTotalElecProducedSA"].quantile(0.25)
q2_unadj_produced = data["unadjustedTotalElecProducedSA"].quantile(0.75)
IQR_move = q2_unadj_produced - q1_unadj_produced
lower_bound = q1_unadj_produced - 1.5 * IQR_move
upper_bound = q2_unadj_produced + 1.5 * IQR_move
```

```python
data = data[(data['unadjustedTotalElecProducedSA']>=upper_bound) | (data['unadjustedTotalElecProducedSA']<=lower_bound)]
```

```python
data.groupby('quarter')['totalElecProducedSA'].mean()
```

```python
data['quarter'].value_counts()
```

```python
plt.figure(figsize = (8,4))
sns.histplot(data = data, x=data['totalElecProducedSA'],
             hue='quarter', alpha=0.5,
             palette = "light:b")
sns.color_palette("viridis", as_cmap=True)
plt.title('Frequency of electricity production volume');
#plt.ylabel('Frequency')
#plt.xlabel('Electricity producton volume');
```

```python
sns.barplot(data=data, x = data['quarter'], y=data['adjElecProdVolume'],
            ci=None);
plt.title('Frequency of electricity produced by Stations');
```

```python
plt.figure(figsize=(10, 5))
ax = sns.lineplot(y=data['totalElecProducedSA'], x=data['Date'], color='salmon', label='SeasonalAdj');
ax = sns.lineplot(y=data['unadjustedTotalElecProducedSA'], x=data['Date'], color='lightblue', label='Unadjusted')
#ax.legend(['Imports', 'Exports']);
plt.ylabel('Electricity')
plt.title('Seasonal adjusted vs Unadjusted');
```
